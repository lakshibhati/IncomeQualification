## Check whether all members of the house have the same poverty level.

#idhogar is unique identifiere for each house hold. Hence for house hold level analysis this feature will be consider

1 Extreme poverty
2 moderate poverty
3 vunarable house hold
4 non vunerable house hold

```python
# house holde with same poverty level
poverty_equal = df_train.groupby('idhogar')['Target'].apply(lambda X:X.unique()==1)

# house holde with unequal poverty level
poverty_unequal = poverty_equal[poverty_equal!=True]

print("The household where all the family members of the house do not have same poverty level", len(poverty_unequal))
```
